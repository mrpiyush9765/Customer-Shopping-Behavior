```python
df['review_rating'] = (
    df.groupby('category')['review_rating']
      .transform(lambda x: x.fillna(x.median()))
)
```

```python
df.isnull().sum()
```

```python
#for apply snake casing replace space in '_',and change upper case to lowercase
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.rename(
    columns={'purchase_amount_(usd)': 'purchase_amount'},
    inplace=True
)
```

```python
df.columns
```

```python
#create a new column age_group
labels=['Young Adult','Middle-aged','Senior']
df['age_group']=pd.qcut(df['age'],q=3,
                        labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
#create column purchase_frequency_days
frequency_mapping = {
    'Weekly': 7,
    'Bi-Weekly': 14,      # if your data uses a hyphen
    'Bi_Weekly': 14,      # if your data uses an underscore
    'Fortnightly': 14,
    'Monthly': 30,
    'Quarterly': 90,
    'Every 3 Months': 90,
    'Annually': 365
}

# Create the new numeric column
df['purchase_frequency_days'] = (
    df['frequency_of_purchases']
      .map(frequency_mapping)
)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
cols = ['discount_applied', 'promo_code_used']
existing_cols = [col for col in cols if col in df.columns]

df[existing_cols].head(10)
```

```python
if 'promo_code_used' in df.columns:
    print((df['discount_applied'] == df['promo_code_used']).all())
else:
    print("promo_code_used column has been removed.")
```

```python
if 'promo_code_used' in df.columns:
    df.drop(columns=['promo_code_used'], inplace=True)
```
